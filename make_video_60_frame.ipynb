```python
video_path = "./video_with_abstacle/TrainingSet_2/tt/"
save_path = "./video_with_abstacle/TrainingSet_2/video_another_30/"

file_list = os.listdir(video_path)

for index, data in enumerate(file_list):
    file_list[index] = video_path + data

file_name = 0
for index, file_path in enumerate(file_list):
    print(file_path)
    cap = cv2.VideoCapture(file_path)
    length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    _60_counter = 0
    
    frame_data = []
    
    for i in range(length):
        ret, frame = cap.read()
        if(ret == True):
            frame = cv2.resize(frame, (356, 356))
            #print(np.shape(frame))
            frame_data.append(frame)
            
            fourcc = cv2.VideoWriter_fourcc(*'XVID')
            out = cv2.VideoWriter(save_path + "%d"%file_name + ".avi", fourcc, 20.0, (356, 356))
            cv2.imshow("video",frame)
            if(_60_counter == 30):
                _60_counter = 0
                file_name = file_name + 1
                print(file_name)
                for j in (frame_data):
                    #print(j)
                    out.write(j)
                frame_data = []
            _60_counter = _60_counter + 1
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            continue
    out.release()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
